## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_____________________________________________________________________________________________________________________
- fault tolerant database is protected from data loss or providing incosistent information by distributing data among multiple servers. The servers contain each other's backups in case of server failure. 
- this also speeds up query times by distributing workload to more servers
- RDD is a collection of such server clusters








## Question 5

Write your answer below this line: 
______________________________________________________________________________________________________________________
