### Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use('seaborn')



df = pd.read_csv('advertising.csv')
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset
# the dataset contains 200 lines/campaings and their resulting sales. The mean sale for all campaings is 14.
#TV is the channel with the highers mean spending (147). Far behind are the other channels - newspaper(30) and radio(23)
# The most successful campaing resulted in sale of 27.
```

### Step 2: Plot histograms with kde overlay to check for the normality of the predictors

```python
# For all the variables, check if they hold normality assumption
tv = df['TV']
radio = df['radio']
news = df['newspaper']
sales = df['sales']

tv.plot.kde(color='navy')
radio.plot.kde(color='pink')
news.plot.kde(color='orange')
plt.legend()
```

```python
# All distributions are fairly normal, although not perfectly symmetrical
# The newspaper distribution has positive skew

```

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries

# build the formula 

# create a fitted model in one line
f ="sales~tv"
model = smf.ols(formula=f,data=df).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"

```python
# the R-squared of this model is 0.612, which is fairly low. This means that a linear regression model would
# not explain the relationship between the variables well.
```


Note here that the coefficients represent associations, not causations
